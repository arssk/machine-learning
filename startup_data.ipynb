```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[::,-1]
y
```

```python
import seaborn as sns
sns.regplot(x=df['R&D Spend'],y=y)
```

```python
sns.regplot(x=df['Administration'],y=y)
```

```python
sns.regplot(x=df['Marketing Spend'],y=y)
```

```python
from sklearn.compose import make_column_transformer      #used to expand as column
from sklearn.preprocessing import OneHotEncoder
one=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=one.fit_transform(x)
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print("percent error is ",mean_absolute_percentage_error(y_test,y_pred))
print("r2_score id",r2_score(y_test,y_pred))
```
