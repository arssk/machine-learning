```python
#the column can be deleted becouse it may cause overfitting
#new encoding method(if we use one hot encoding ,there is a chance of increasing vlues and get overfitted)
df5=pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type']],drop_first=True)
df5
```

```python
#now concatinating with the two dataframe
dfe=pd.concat([df,df5],axis=1)
dfe
```

```python
dfe.columns
```

```python
#now droping unnecessory columns
#unnamed,name,location,fueltype,transmissionn,owner type,neew price(unnessocery)
dfe.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price'],axis=1,inplace=True)
dfe
```

```python
#noe there is some unit along with the data.so we have to remove that object type in the data
dfe['Mileage']=dfe['Mileage'].str.replace('km/kg','')
dfe
```

```python
dfe['Mileage']=dfe['Mileage'].str.replace('kmpl','')
dfe
```

```python
dfe['Engine']=dfe['Engine'].str.replace('CC','')
dfe
```

```python
dfe['Power']=dfe['Power'].str.replace('bhp','')
dfe
```

```python
dfe.dtypes
```

```python
#at some place in emgine and power ther is a variable typed as null.so that also should be changed too
#replaced with '0'
dfe['Engine']=dfe['Engine'].str.replace('null','0')
dfe['Power']=dfe['Power'].str.replace('null','0')
dfe['Mileage']=dfe['Mileage'].str.replace('null','0')
dfe
```

```python
#now converting the data type of the objest columns(mileage,engine,power)
dfe['Engine']=dfe['Engine'].astype(float)
dfe['Mileage']=dfe['Mileage'].astype(float)
dfe['Power']=dfe['Power'].astype(float)
dfe.dtypes
```

```python
#now the places that had null now have a value  of zero.but no engine would never have value of zero.so now it will be treated as a miissing value
#and repllaced with numpy(becouse the alue is a numarical)
dfe.loc[dfe.Engine==0,'Engine']=np.NAN
dfe.loc[dfe.Mileage==0,'Mileage']=np.NAN
dfe.loc[dfe.Power==0,'Power']=np.NAN
```

```python
dfe.isna().sum()
```

```python
dfe['Mileage']=dfe['Mileage'].fillna(dfe['Mileage'].mean())
dfe['Engine']=dfe['Engine'].fillna(dfe['Engine'].mean())
dfe['Power']=dfe['Power'].fillna(dfe['Power'].mean())
dfe['Seats']=dfe['Seats'].fillna(dfe['Seats'].mode()[0])
dfe
```
